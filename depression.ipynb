```python
def convert_sleep_dur(x):
    res = []
    for ind, sample in enumerate(x):
        res.append([])
        sample = sample.replace("-", " ").split()
        for ltr in sample:
            if ltr.isdigit() and int(ltr) < 13: 
                res[ind].append(int(ltr))
        res[ind].append(0)
        res[ind] = max(res[ind])
    mean_hour = np.mean(res)
    res = [mean_hour if i == 0 else i for i in res]
    return res

print(convert_sleep_dur(data["Sleep Duration"])[:10])
print(data["Sleep Duration"][:10])
```

```python
data["Working Professional or Student"] = data["Working Professional or Student"].apply(lambda x: 1 if x == "Working Professional" else 0)
data["Dietary Habits"] = data["Dietary Habits"].apply(lambda x: 1 if x == "Healthy" else 0)
data["Gender"] = data["Gender"].apply(lambda x: 1 if x == "Male" else 0)
data["Family History of Mental Illness"] = data["Family History of Mental Illness"].apply(lambda x: 1 if x == "Yes" else 0)
data["Have you ever had suicidal thoughts ?"] = data["Have you ever had suicidal thoughts ?"].apply(lambda x: 1 if x == "Yes" else 0)
data["Sleep Duration"] = convert_sleep_dur(data["Sleep Duration"])

num_features = ["Age", "Working Professional or Student", "Dietary Habits", "Academic Pressure", "Work Pressure", "CGPA",
                      "Study Satisfaction", "Job Satisfaction", "Sleep Duration", "Work/Study Hours", "Financial Stress", "Depression",
                      "Family History of Mental Illness", "Gender", "Have you ever had suicidal thoughts ?"]

cat_features = data.columns.difference(num_features)
print("Категориальные признаки:")
print(*list(cat_features))
print("Числовые признаки:")
print(*num_features)
```

```python
pd.get_dummies(data[cat_features])
```

```python
data.corr(numeric_only=True, ).style.background_gradient(cmap='coolwarm')
```

```python
for col in num_features:
    data[col] = data[col].fillna(data[col].mean())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(data.drop(cat_features, axis=1).drop("Depression", axis=1), data["Depression"], test_size=0.3, shuffle=True)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression()
model.fit(X_train, y_train)
accuracy_score(y_test, model.predict(X_test))
```
